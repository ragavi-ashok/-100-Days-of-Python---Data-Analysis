```python
## Website visitors tracking
## Develop a line chart to track the daily number of visitors to a website over the last year, color 'midnightblue' for the line.
## Matplotlib Color Mapping: color='midnightblue'Matplotlib Figure Size: figsize=(14, 7)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {
    "date": [
        "2023-01-01", "2023-01-02", "2023-01-03", "2023-01-04", "2023-01-05"],
    "visitors": [1392, 727, 693, 1808, 1758]
}
```

```python
# Convert the dictionary into a DataFrame
df = pd.DataFrame(data)
```

```python
# Ensure 'date' is in datetime format
df['date'] = pd.to_datetime(df['date'])
```

```python
# Set up the plot
plt.figure(figsize=(14,7))
plt.plot(df['date'], df['visitors'], color='midnightblue')
# Set title and labels
plt.title("Daily Number of Visitors Over the Last Year", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Number of Visitors", fontsize=12)
# Show the plot
plt.grid(True)
plt.show()
```
